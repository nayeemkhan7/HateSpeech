```python
import pandas as pd
import numpy as np
import re
import string
import nltk
```

```python
df = pd.read_csv('D:/Pre/5000Tweet.csv')
```

```python
df
```

```python
del df["count"]
del df['Unnamed: 0']
del df['hate_speech']
del df['offensive_language']
del df['neither']
```

```python
df.pop("class")
```

```python
df
```

```python
# Lowercasing

def lower_tweet(text):
  text = text.lower()
  return text

df['tweet'] = df['tweet'].apply(lower_tweet)
```

```python
df
```

```python
#Remove pattern from the input text

def remove_pattern(input_text, pattern):
  r = re.findall(pattern, input_text)
  for word in r:
    input_text = re.sub(word, "", input_text)
  return input_text
```

```python
#Remove twitter handle (@username)

df['tweet'] = np.vectorize(remove_pattern)(df['tweet'], "@[\w]*")
```

```python
df
```

```python
#Remove special characters:

df['tweet'] = df['tweet'].str.replace("[^a-zA-Z#]", " ")
```

```python
df
```

```python
#Tokenization

df['tweet'] = df['tweet'].apply(lambda x: x.split())
```

```python
df
```

```python
#Import stopwords list

nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python
# Remove stopword from List

def remove_stopwords(tweet):
  tweets = [word for word in tweet if word not in stopwords]
  return tweets

df['tweet'] = df['tweet'].apply(lambda x: remove_stopwords(x))
```

```python
df
```

```python
#Stemming

from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

df['tweet'] = df['tweet'].apply(lambda sentence: [stemmer.stem(word) for word in sentence])
```
